# 1. Create dataframe for neighbourhoods in Toronto

```python
import pandas as pd # library for data analsysis

import requests # library to handle requests
from bs4 import BeautifulSoup
```

```python
#get the url of the page
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
#request the response in the form of html
s = requests.Session()
response = s.get(url, timeout=10)
response
```

```python
#parse data from html with Beautifulsoup
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
#parse the data into BS object
table = soup.find('table',{'class':'wikitable sortable'})
table
```

```python
#convert wiki table into Pandas dataframe
df = pd.read_html(str(table))
#convert list to dataframe
df = pd.DataFrame(df[0])
print(df.head())
df.shape
```

Note: Combining the neighbourhoods with the same postcode into one row is done in the import.

```python
#Delete rows without a borough: Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
df = df[df.Borough != 'Not assigned']
df.shape
```

```python
# Give value to Neighbourhood: If a cell has a borough but a Not assigned  neighborhood, then the neighborhood will be the same as the borough.
for row in range(len(df)):
    if df['Neighbourhood'].iloc[row] =='Not assigned':
        df['Neighbourhood'].iloc[row]= df['Borough'].iloc[row]
```

```python
#shape: In the last cell of your notebook, use the .shape method to print the number of rows of your dataframe.
df.shape
```

# 2. Get coordinates for each neighborhoood

```python
#
import csv
import io
import urllib.request
from pandas import DataFrame

geo_url = "https://cocl.us/Geospatial_data"
webpage = urllib.request.urlopen(geo_url)
geo_reader = csv.reader(webpage.read().decode('utf-8').splitlines())
```

```python
#get geo data into a list
geo_list = []
for row in geo_reader:
    geo_list.append(row)
print(geo_list)
```

```python
#create pandas df for the coordinates
geo_df_headers = geo_list.pop(0)
geo_df = DataFrame (geo_list, columns=geo_df_headers)
geo_df.head()
```

```python
#add the informationto our dataframe
df = pd.merge(df, geo_df, how='left', left_on='Postal Code', right_on='Postal Code')
df.head(10)
```

# 3. Analyze Each Neighborhood

```python
#import packages we might need
import numpy as np # library to handle data in a vectorized manner

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
import matplotlib.pyplot as plt

# import k-means from clustering stage
from sklearn.cluster import KMeans

print('Libraries imported.')
```

```python
#Get Toronto Canda coordinates
address = "Toronto, Canada"

latitude_to = '43.6532'
longitude_to = '-79.3832'
print('The geograpical coordinate of Toronto, Canada are {}, {}.'.format(latitude_to, longitude_to))
```

Create a map of Toronto with Neighborhoods superimposed on top

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude_to, longitude_to], zoom_start=10)

#create new dataframe for Toronto
toronto_df = df

# add markers to map
for lat, lng, borough, neighborhood in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Borough'], toronto_df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

### Cluster Neighbourhoods

Cluster the neighbourhoods into 5 clusters

```python
# set number of clusters
kclusters = 5

toronto_clustering = toronto_df.filter(['Latitude','Longitude'], axis=1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Visualise each location with the respective cluster

```python
#add clustering labels
toronto_df.insert(0, 'Cluster Labels', kmeans.labels_)

# create map
map_clusters = folium.Map(location=[latitude_to, longitude_to], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Neighbourhood'], toronto_df['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
